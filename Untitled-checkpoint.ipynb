## Data Preprocessing

```python
def preprocess_data(df, is_train=True, label_encoders=None):

    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]  # Drop unnamed cols

    if is_train:
        label_encoders = {}

        for col in df.select_dtypes(include='object').columns:
            le = LabelEncoder()
            df[col] = le.fit_transform(df[col].astype(str))
            label_encoders[col] = le
    else:
        if label_encoders is None:
            raise ValueError("label_encoders must be provided for test/inference.")

        for col in df.select_dtypes(include='object').columns:
            if col in label_encoders:
                le = label_encoders[col]
                df[col] = le.transform(df[col].astype(str))
            else:
                raise ValueError(f"Missing encoder for column: {col}")

    df = df.fillna(df.mean(numeric_only=True))

    if is_train:
        X = df.drop(columns=['Engagement_Score', 'Performance_Score'])
        y = df['Engagement_Score']
        y1 = df['Performance_Score']
        return X, y, y1, label_encoders
    else:
        return df
```

```python
X,y,y1, label_s  = preprocess_data(df,is_train=True)
```

```python
def perform_eda(X, y, y1):
    """
    Perform Exploratory Data Analysis for the music education dataset.

    Parameters:
    - X: DataFrame of features
    - y_reg: DataFrame with ['Performance_Score', 'Engagement_Score']
    """
    plt.figure(figsize=(18, 12))

    # 1. Count Plot: Class_Level distribution
    plt.subplot(2, 3, 1)
    sns.countplot(x=X['Class_Level'])
    plt.title("Distribution of Class Levels")
    plt.xlabel("Class Level")
    plt.ylabel("Count")

    # 2. Box Plot: Heart Rate vs Performance Score
    plt.subplot(2, 3, 2)
    sns.boxplot(x=pd.cut(X['Heart_Rate (BPM)'], bins=5), y=y1)
    plt.title("Performance Score vs Heart Rate")
    plt.xlabel("Heart Rate Range")
    plt.ylabel("Performance Score")

    # 3. Violin Plot: Engagement Level vs Engagement Score
    plt.subplot(2, 3, 3)
    sns.violinplot(x=X['Engagement_Level'], y=y)
    plt.title("Engagement Score by Engagement Level")
    plt.xlabel("Engagement Level (1–10)")
    plt.ylabel("Engagement Score")

    # 4. Box Plot: Pitch Accuracy vs Performance Score
    plt.subplot(2, 3, 4)
    sns.boxplot(x=pd.cut(X['Pitch_Accuracy'], bins=5), y=y1)
    plt.title("Performance Score vs Pitch Accuracy")
    plt.xlabel("Pitch Accuracy Range")
    plt.ylabel("Performance Score")

    # 5. Scatter Plot: Focus Time vs Engagement Score
    plt.subplot(2, 3, 5)
    sns.scatterplot(x=X['Focus_Time (min)'], y=y)
    plt.title("Engagement Score vs Focus Time")
    plt.xlabel("Focus Time (seconds)")
    plt.ylabel("Engagement Score")

    # 6. Correlation Heatmap
    plt.subplot(2, 3, 6)
    corr_df = X.copy()
    corr_df['Engagement_Score'] = y
    corr_df['Performance_Score'] = y1
    sns.heatmap(corr_df.corr(), annot=False, cmap='coolwarm')
    plt.title("Correlation Heatmap")
    
    
    plt.tight_layout()
    plt.savefig("results/music_eda_plots.png")
    plt.show()
```

## Evaluate Metrics

```python
# Lists to store metrics for all models
mae_list = []
mse_list = []
rmse_list = []
r2_list = []
model_names = []
results = {}

def evaluate_performance(y_true, y_pred, model_name):
    """Evaluate model performance, plot predictions, and save scatterplot."""
    # Calculate metrics
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    
    # Append metrics to lists
    model_names.append(model_name)
    mae_list.append(mae)
    mse_list.append(mse)
    rmse_list.append(rmse)
    r2_list.append(r2)
    
    # Print metrics
    print(f"\n{model_name} Performance:")
    print(f"MAE: {mae:.4f}")
    print(f"MSE: {mse:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"R2 score: {r2:.4f}")
    
    # Create scatterplot
    plt.figure(figsize=(10, 6))
    plt.scatter(y_true, y_pred, alpha=0.5)
    plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--', lw=2)
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.title(f'Actual vs Predicted Values - {model_name}')
    
    # Create models directory if it doesn't exist
    os.makedirs('results', exist_ok=True)
    
    # Save plot
    plot_path = os.path.join('results', f'{model_name}_scatterplot.png')
    plt.savefig(plot_path)
    plt.show()
    
    return {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R2 score': r2}

def create_comparison_table():
    """Create and display a comparison table of all models' performance metrics."""
    # Create DataFrame from collected metrics
    comparison_df = pd.DataFrame({
        'Model': model_names,
        'MAE': mae_list,
        'MSE': mse_list,
        'RMSE': rmse_list,
        'R2 Score': r2_list
    })
    
    # Round numeric columns to 4 decimal places
    numeric_columns = ['MAE', 'MSE', 'RMSE', 'R2 Score']
    comparison_df[numeric_columns] = comparison_df[numeric_columns].round(4)

    
    # Save comparison table as CSV
    os.makedirs('results', exist_ok=True)
    comparison_df.to_csv('results/regression_model_comparison.csv', index=False)
    
    return comparison_df
```

## KNN Regressor

```python
from sklearn.neighbors import KNeighborsRegressor
from joblib import dump, load
import os
import matplotlib.pyplot as plt

def train_knn_regressor(X_train, y_train, X_test, y_test):
    """Train and evaluate KNN Regressor model."""
    model_path = os.path.join(MODEL_DIR, 'knn_regressor.joblib')
    
    if os.path.exists(model_path):
        model = load(model_path)
        print("Loaded KNN Regressor model.")
    else:
        model = KNeighborsRegressor(n_neighbors=5)
        model.fit(X_train, y_train)
        dump(model, model_path)
        print("Trained and saved new KNN Regressor model.")
    
    y_pred = model.predict(X_test)

    return evaluate_performance(y_test, y_pred, "KNN Regressor")
```

```python
results['KNN Regressor'] = train_knn_regressor(X_train, y_eng_train, X_test, y_eng_test)
```

## Support Vector Regressor

```python
from sklearn.svm import SVR

def train_svr(X_train, y_train, X_test, y_test):
    """Train and evaluate SVR model."""
    model_path = os.path.join(MODEL_DIR, 'svr.joblib')
    
    if os.path.exists(model_path):
        model = load(model_path)
        print("Loaded Support Vector Regression model.")
    else:
        model = SVR(kernel='rbf')
        model.fit(X_train, y_train)
        dump(model, model_path)
        print("Trained and saved new Support Vector Regression model.")
    
    y_pred = model.predict(X_test)

    return evaluate_performance(y_test, y_pred, "SVR Model")
```

```python
results['SVR'] = train_svr(X_train, y_eng_train, X_test, y_eng_test)
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from joblib import dump, load
import os
import matplotlib.pyplot as plt

def train_decision_tree_regressor(X_train, y_train, X_test, y_test):
    """Train and evaluate Decision Tree Regressor model."""
    model_path = os.path.join(MODEL_DIR, 'decision_tree_regressor.joblib')
    
    if os.path.exists(model_path):
        model = load(model_path)
        print("Loaded Decision Tree Regressor model.")
    else:
        model = DecisionTreeRegressor(max_depth=2, random_state=42)
        model.fit(X_train, y_train)
        dump(model, model_path)
        print("Trained and saved new Decision Tree Regressor model.")
    
    y_pred = model.predict(X_test)

    return evaluate_performance(y_test, y_pred, "Decision Tree Regressor")
```

```python
results['DTR'] = train_decision_tree_regressor(X_train, y_eng_train, X_test, y_eng_test)
```

## Axis Forest Regressor (Hybrid Model)

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, StackingRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from joblib import dump, load
import os

def train_hybrid_regressor(X_train, y_train, X_test, y_test):
    """Train and evaluate hybrid model: Linear Regression (base) + Random Forest (meta)."""
    model_path = os.path.join(MODEL_DIR, 'hybrid_regressor.joblib')

    if os.path.exists(model_path):
        model = load(model_path)
        print("Loaded Hybrid Regressor model.")
    else:
        base_model = LinearRegression()
        meta_model = RandomForestRegressor(n_estimators=100, random_state=42)

        model = StackingRegressor(
            estimators=[('lr', base_model)],
            final_estimator=meta_model,
            passthrough=True  # Optionally allow original features to be used by the meta model
        )
        model.fit(X_train, y_train)
        dump(model, model_path)
        print("Trained and saved new Hybrid Regressor model.")

    y_pred = model.predict(X_test)
    
    return evaluate_performance(y_test, y_pred, "AxisForest")
```

```python
results['Hybrid'] = train_hybrid_regressor(X_train, y_eng_train, X_test, y_eng_test)
```

```python
def predict_scores_on_test_data(test_df, label_encoders):
    """Predict Engagement_Score and Performance_Score on new test data."""
    # Preprocess test data
    processed_df = preprocess_data(test_df, is_train=False, label_encoders=label_encoders)

    # Load saved models
    engagement_model_path = os.path.join(MODEL_DIR, 'hybrid_regressor.joblib')
    performance_model_path = os.path.join(MODEL_DIR, 'hybrid_lasso_rf_regressor.joblib')

    if not os.path.exists(engagement_model_path) or not os.path.exists(performance_model_path):
        raise FileNotFoundError("Required models not found in 'models/' directory.")

    engagement_model = load(engagement_model_path)
    performance_model = load(performance_model_path)

    # Predict scores
    engagement_preds = engagement_model.predict(processed_df)
    performance_preds = performance_model.predict(processed_df)

    # Add predictions to original test data
    test_df['Predicted_Engagement_Score'] = engagement_preds
    test_df['Predicted_Performance_Score'] = performance_preds

    # Save predictions
    output_path = os.path.join(RESULTS_DIR, 'test_predictions.csv')
    test_df.to_csv(output_path, index=False)
    print(f"\n Predictions saved to: {output_path}")

    return test_df
```
